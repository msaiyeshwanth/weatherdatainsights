### Merging the datasets

```python
melted_humidity_df = pd.melt(humidity_df, id_vars=['datetime'], var_name='City', value_name='humidity')
melted_pressure_df = pd.melt(pressure_df, id_vars=['datetime'], var_name='City', value_name='pressure')
melted_temperature_df = pd.melt(temperature_df, id_vars=['datetime'], var_name='City', value_name='temperature')
melted_desc_df = pd.melt(desc_df, id_vars=['datetime'], var_name='City', value_name='weather_description')
melted_windspeed_df = pd.melt(windspeed_df, id_vars=['datetime'], var_name='City', value_name='wind_speed')
melted_direction_df = pd.melt(direction_df, id_vars=['datetime'], var_name='City', value_name='wind_direction')
```

```python
merged_df = pd.merge(melted_humidity_df, melted_pressure_df, how='inner', on=['datetime', 'City'])
merged_df = pd.merge(merged_df, melted_temperature_df, how='inner', on=['datetime', 'City'])
merged_df = pd.merge(merged_df, melted_desc_df, how='inner', on=['datetime', 'City'])
merged_df = pd.merge(merged_df, melted_windspeed_df, how='inner', on=['datetime', 'City'])
merged_df = pd.merge(merged_df, melted_direction_df, how='inner', on=['datetime', 'City'])
merged_df.head()
```

```python
df = pd.merge(merged_df, city_df, how='inner', on='City')
df.head()
```

```python
# checking the dimensions of the data
df.shape
```

# Data Cleaning

### Sampling the data

```python
# checking the range of datetime
df.datetime.describe()
```

```python
# sampling 2017 US data as the data is huge
df['datetime'] = pd.to_datetime(df['datetime'])

sample_df = df[(df['datetime'].dt.year == 2017) & (df['Country'] == 'United States')]

sample_df = sample_df.reset_index(drop=True)

sample_df.head()
```

```python
# dropping Country as it has only US
sample_df = sample_df.drop('Country', axis=1)
sample_df.head()
```

### Checking Column names consistency

```python
# checking column names
sample_df.columns
```

### Cleaning numerical columns i.e, 'humidity', 'pressure', 'temperature', 'wind_speed', 'wind_direction'

```python
# checking summary statistics for numerical columns
sample_df.describe()
```

```python
# using ffill() to impute previous values from the same day.
numeric_columns = ['humidity', 'pressure', 'temperature', 'wind_speed', 'wind_direction']

sample_df[numeric_columns] = sample_df.groupby([sample_df['datetime'].dt.date, 'city'])[numeric_columns].ffill()

sample_df.isna().sum()
```

```python
# checking summary statistics for numerical columns
sample_df.describe()
```

```python
# imputing mean for missing values for which ffil() did not work.
sample_df[numeric_columns] = sample_df.groupby([sample_df['datetime'].dt.date, 'city'])[numeric_columns].transform(lambda x: x.fillna(x.mean()))
sample_df.isna().sum()
```

```python
# checking summary statistics for numerical columns
sample_df.describe()
```

### Cleaning Temperature Column

```python
# Converting temperature from Kelvin to Farenheit

sample_df['temperature'] = ((sample_df['temperature'] - 273.15) * 9/5) + 32

sample_df.temperature.describe()
```

### Creating Label

```python
sample_df.weather_description.value_counts()
```

```python
# creating a function to label data based on weather description
def weather_labels(description):
    if 'clear' in description:
        return 'clear'
    elif 'rain' in description or 'drizzle' in description:
        return 'rainy'
    elif 'snow' in description:
        return 'snowy'
    elif 'thunderstorm' in description or 'thunder' in description:
        return 'thunderstorm'
    elif 'mist' in description or 'fog' in description or 'haze' in description or 'smoke' in description:
        return 'foggy'
    elif 'clouds' in description:
        return 'cloudy'
    else:
        return 'other'
```

```python
# converting weather description to lower case 
sample_df['weather_description'] = sample_df['weather_description'].str.lower()

# applying weather_labels function
sample_df['weather'] = sample_df['weather_description'].apply(weather_labels)
sample_df.weather.value_counts()
```

```python
sample_df.shape
```

```python
sample_df = sample_df.drop(columns=['weather_description'])
sample_df = sample_df.reset_index(drop=True)
sample_df.head()
```

### 3. Which cities had the most days with a particular weather condition?

```python
weather_labels = cleaned_df['weather'].unique()

for label in weather_labels:
    subset_df = cleaned_df[cleaned_df['weather'] == label]
    
    city_counts = subset_df['city'].value_counts().sort_values(ascending=False).head(5)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=city_counts.values, y=city_counts.index, palette='Blues_r')
    plt.title(f'Top 5 Cities with Most {label.capitalize()} Days')
    plt.xlabel('Count')
    plt.ylabel('City')
    plt.show()
```

### Interpretation:

Foggy - San Diego

Rainy - Seattle

Snowy - Pittsburgh

Cloudy - Albuquerque

Clear - Las Vegas

Thunderstorm - Miami

Other - Los Angeles

### 4. How do temperature and humidity vary over different time intervals in Denver (2017) within the dataset, and are there noticeable trends or anomalies?

```python
denver_df = cleaned_df[cleaned_df['city'] == 'Denver']
denver_df['month'] = cleaned_df['datetime'].dt.month_name()

denver_df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='temperature', data = denver_df, label='Temperature')
sns.lineplot(x='month', y='humidity', data = denver_df, label='Humidity')
plt.title('Time Series Plot of Temperature and Humidity in Denver (2017)')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### Interpretation:
1. Temperature starts to increase from January to July and then decreases gradually from July to November.
2. January is the coldest month and July is the hottest month.
3. There is no significant pattern in Humidity.
4. January and May recorded highest humidity and March recorded Lowest Humidity.
