```python
import pandas as pd
import requests
import os
from datetime import datetime
```

```python
api_key = ''
cities = ['Portland', 'Kansas City', 'Philadelphia', 'Pittsburgh', 'Charlotte', 'Jacksonville', 'Detroit', 'Atlanta', 'Indianapolis', 'Nashville', 'Chicago', 'Saint Louis', 'Minneapolis', 'Houston', 'Dallas', 'San Antonio', 'Denver', 'Albuquerque', 'Phoenix', 'Las Vegas', 'San Diego', 'Los Angeles', 'Seattle', 'Boston', 'San Francisco', 'Miami', 'New York']

city_data = []

for city in cities:
    url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()

        city_weather_data = {
            'datetime': pd.to_datetime('now'),
            'city': city,
            'humidity': data['main']['humidity'],
            'pressure': data['main']['pressure'],
            'temperature': data['main']['temp'],
            'weather_description': data['weather'][0]['description'],
            'wind_speed': data['wind']['speed'],
            'wind_direction': data['wind']['deg'],
            'latitude': data['coord']['lat'],
            'longitude': data['coord']['lon']
        }
        city_data.append(city_weather_data)
        print(f"Weather data for {city} collected.")
    else:
        print(f"Failed to fetch weather data for {city}. Status code: {response.status_code}")

df = pd.DataFrame(city_data)
df.head()
```

```python
df.dtypes
```

```python
df['datetime'] = df['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
df.head()
```

```python
df['temperature'] = ((df['temperature'] - 273.15) * 9/5) + 32

df.head()
```

```python
df.to_csv('weatherapidata.csv',index=False)
```
